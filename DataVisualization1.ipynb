#### Importing all necessary modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Loading the database

```python
gas_prices_db = pd.read_csv("gas_prices.csv")
print("Shape:",gas_prices_db.shape,end='\t\t')
print("Size:",gas_prices_db.size)
gas_prices_db.sample(5)
```

```python
gas_prices_db.describe()
```

## Histogram

```python
plt.figure(figsize=(5,4),dpi=100)
plt.hist(gas_prices_db.Japan,bins=[2,3,4,5,6],label='Japan',color='#f05e3a')
plt.xlabel("Price of Gas in Dollars")
plt.ylabel("Frequency")
plt.title("Gas Price frequency chart",fontdict={'fontsize':15,'fontweight':'bold'})
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,8))
clist = ['Japan','USA','Canada','UK']
color = ['#092327','#0b5351','#00a9a5','#27fb6b']
for index,country in enumerate(clist):
    plt.subplot(2,2,index+1)
    plt.hist(gas_prices_db[country],bins=[1,2,3,4,5,6,7,8],label=country,color=color[index])
    plt.xlabel("Price of Gas in Dollars")
    plt.ylabel("Frequency")
    plt.xticks([i for i in range(1,9)])
    plt.legend()
plt.savefig("GasPrice.png",dpi=400)
plt.show()
```

```python
plt.figure(figsize=(20,8))
clist = list(gas_prices_db.columns[1:])
def gen_random_color():
    return '#{:06x}'.format(np.random.randint(0, 256**3))
color = [gen_random_color() for i in range(len(clist))]
for index,country in enumerate(clist):
    plt.subplot(2,int(np.ceil(len(clist)/2)),index+1)
    plt.hist(gas_prices_db[country],bins=[1,2,3,4,5,6,7,8],label=country,color=color[index])
    plt.xlabel("Price of Gas in Dollars")
    plt.ylabel("Frequency")
    plt.xticks([i for i in range(1,9)])
    plt.legend()
plt.savefig("GasPrice2.png",dpi=500)
plt.show()
```

## Line plot

```python
plt.style.use('tableau-colorblind10')
plt.plot(gas_prices_db.Year,gas_prices_db.Japan,color='#EF6351',label='Japan')
plt.plot(gas_prices_db.Year,gas_prices_db.Canada,color='#F38375',label='Canada')
plt.plot(gas_prices_db.Year,gas_prices_db.USA,color='#49416D',label='USA')
plt.plot(gas_prices_db.Year,gas_prices_db.France,color='#A882DD',label='France')
plt.xticks([i for i in range(1990,2009,2)])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
clist = np.random.choice(list(gas_prices_db.columns[1:]),size=6)
def gen_random_color():
    return '#{:06x}'.format(np.random.randint(0, 256**3))
color = [gen_random_color() for i in range(len(clist))]
for index,country in enumerate(clist):
    plt.plot(gas_prices_db.Year,gas_prices_db[country],color=color[index],label=country)
plt.xticks([i for i in range(1990,2009,2)])
plt.xlabel("Years")
plt.ylabel("Price in Dollar")
plt.title("Price of Gas ($) in random countries 1990-2008",fontdict={'fontsize':20,'fontweight':'bold'})
plt.legend()
plt.show()
```
